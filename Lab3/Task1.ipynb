
Задача 7
Фірма планує виробництво нової продукції швидкого харчування в національному масштабі. Дослідницький відділ переконаний у великому успіху нової продукції і хоче впровадити її негайно, без рекламної кампанії на ринках збуту фірми. Відділ маркетингу стан речей оцінює інакше і пропонує провести інтенсивну рекламну кампанію. Така кампанія обійдеться в 100 тис. дол., а в разі успіху принесе 950 тис. дол. річного доходу. У разі провалу рекламної кампанії (ймовірність цього становить 30%) річний дохід оцінюється лише в 200 тис. дол. Якщо рекламна кампанія не проводиться зовсім, річний дохід оцінюється в 400 тис. дол. за умови, що покупцям сподобається нова продукція (ймовірність цього дорівнює 0,8), і в 	200 тис. дол. з імовірністю 0,2, якщо покупці залишаться байдужими до нової продукції.
Побудуйте відповідне дерево рішень. Як повинна вчинити фірма у зв'язку з виробництвом нової продукції?


```python
import numpy as np
import matplotlib.pyplot as plt

# Ймовірності подій
p_success_ad = 0.7
p_failure_ad = 0.3
p_success_no_ad = 0.8
p_neutral_no_ad = 0.2

# Вартість різних можливих подій
revenue_ad_success = 950
revenue_ad_failure = 200
revenue_no_ad_success = 400
revenue_no_ad_neutral = 200

# Вартість проведення рекламної кампанії
cost_ad = 100

# Обчислення вартостей для різних стратегій
expected_value_ad = p_success_ad * (revenue_ad_success - cost_ad) + p_failure_ad * (revenue_ad_failure - cost_ad)
expected_value_no_ad = p_success_no_ad * revenue_no_ad_success + p_neutral_no_ad * revenue_no_ad_neutral

# Порівняння стратегій
if expected_value_ad > expected_value_no_ad:
    decision = "Провести рекламну кампанію"
else:
    decision = "Не проводити рекламну кампанію"

# Вивід результату
print(f"Очікувана вартість з рекламною кампанією: {expected_value_ad} тис. дол.")
print(f"Очікувана вартість без рекламної кампанії: {expected_value_no_ad} тис. дол.")
print(f"Рекомендація фірмі: {decision}")

```

```python
# Функція корисності (можна модифікувати)
def utility_function(wealth, risk):
    return wealth - risk * wealth

# Діапазон грошових сум
wealth_range = np.arange(0, 1001, 10)

# Рівні ризику
risks = [0.1, 0.5, 0.9]

# Побудова графіка для кожного рівня ризику
for risk in risks:
    utility_values = [utility_function(wealth, risk) for wealth in wealth_range]
    plt.plot(wealth_range, utility_values, label=f'Ризик {risk}')

# Налаштування графіка
plt.xlabel('Грошові суми')
plt.ylabel('Функція корисності')
plt.legend()
plt.title('Графік функції корисності при різних рівнях ризику')

# Показ графіка
plt.grid(True)
plt.show()
```

Пряма для ризику 0.1:
Для низьких рівнів ризику (0.1) корисність практично не зменшується зі збільшенням грошових сум. Тобто, навіть при великих сумах грошей ризик має дуже маленький вплив на сприйняття корисності.
Ця пряма відображає ставлення до ризику, де людина має високий рівень терпимості до ризику та відчувається комфортно при великих інвестиціях з маленькими коливаннями.

Пряма для ризику 0.5:
Для нейтрального рівню ризику (0.5) корисність зменшується повільніше, аніж для високого ризику, при збільшенні грошових сум.
Ця пряма відображає середнє ставлення до ризику, де людина має помірний рівень терпимості до ризику та може приймати помірні інвестиційні ризики.

Пряма для ризику 0.9:
Для високих рівнів ризику (0.9) корисність зменшується швидше зі збільшенням грошових сум.
Ця пряма відображає ставлення до ризику, де людина має низький рівень терпимості до ризику та може бути готовою приймати значні ризики для потенційно високого доходу.
