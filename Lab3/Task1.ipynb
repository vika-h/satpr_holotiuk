
Задача 7
Фірма планує виробництво нової продукції швидкого харчування в національному масштабі. Дослідницький відділ переконаний у великому успіху нової продукції і хоче впровадити її негайно, без рекламної кампанії на ринках збуту фірми. Відділ маркетингу стан речей оцінює інакше і пропонує провести інтенсивну рекламну кампанію. Така кампанія обійдеться в 100 тис. дол., а в разі успіху принесе 950 тис. дол. річного доходу. У разі провалу рекламної кампанії (ймовірність цього становить 30%) річний дохід оцінюється лише в 200 тис. дол. Якщо рекламна кампанія не проводиться зовсім, річний дохід оцінюється в 400 тис. дол. за умови, що покупцям сподобається нова продукція (ймовірність цього дорівнює 0,8), і в 	200 тис. дол. з імовірністю 0,2, якщо покупці залишаться байдужими до нової продукції.
Побудуйте відповідне дерево рішень. Як повинна вчинити фірма у зв'язку з виробництвом нової продукції?


```python
# Ймовірності подій
p_success_ad = 0.7
p_failure_ad = 0.3
p_success_no_ad = 0.8
p_neutral_no_ad = 0.2

# Вартість різних можливих подій
revenue_ad_success = 950
revenue_ad_failure = 200
revenue_no_ad_success = 400
revenue_no_ad_neutral = 200

# Вартість проведення рекламної кампанії
cost_ad = 100

# Обчислення вартостей для різних стратегій
expected_value_ad = p_success_ad * revenue_ad_success + p_failure_ad * revenue_ad_failure - cost_ad
expected_value_no_ad = p_success_no_ad * revenue_no_ad_success + p_neutral_no_ad * revenue_no_ad_neutral

# Порівняння стратегій
if expected_value_ad > expected_value_no_ad:
    decision = "Провести рекламну кампанію"
else:
    decision = "Не проводити рекламну кампанію"

# Вивід результату
print(f"Очікувана вартість з рекламною кампанією: {expected_value_ad} тис. дол.")
print(f"Очікувана вартість без рекламної кампанії: {expected_value_no_ad} тис. дол.")
print(f"Рекомендація фірмі: {decision}")

```

Аналіз чутливості отриманого рішення

```python
def sensitivity_analysis(p_success_ad, p_failure_ad, p_success_no_ad, p_neutral_no_ad, 
                        revenue_ad_success, revenue_ad_failure, revenue_no_ad_success, revenue_no_ad_neutral, 
                        cost_ad):
    results = []

    for p_success_ad_new in [0.3, 0.6, 0.8]:  # Перевіряємо різні значення ймовірності успіху рекламної кампанії
        for revenue_ad_success_new in [900, 950, 1000]:  # Перевіряємо різні значення вартості успішної рекламної події
            expected_value_ad = p_success_ad_new * (revenue_ad_success_new - cost_ad) + (1 - p_success_ad_new) * (revenue_ad_failure - cost_ad)

            expected_value_no_ad = p_success_no_ad * revenue_no_ad_success + p_neutral_no_ad * revenue_no_ad_neutral

            if expected_value_ad > expected_value_no_ad:
                decision = "Провести рекламну кампанію"
            else:
                decision = "Не проводити рекламну кампанію"

            results.append((p_success_ad_new, revenue_ad_success_new, expected_value_ad, decision))

    return results

# Вхідні параметри для аналізу чутливості
p_success_ad = 0.7
p_failure_ad = 0.3
p_success_no_ad = 0.8
p_neutral_no_ad = 0.2
revenue_ad_success = 950
revenue_ad_failure = 200
revenue_no_ad_success = 400
revenue_no_ad_neutral = 200
cost_ad = 100

sensitivity_results = sensitivity_analysis(p_success_ad, p_failure_ad, p_success_no_ad, p_neutral_no_ad, 
                                          revenue_ad_success, revenue_ad_failure, revenue_no_ad_success, revenue_no_ad_neutral, 
                                          cost_ad)

for p_success_ad_new, revenue_ad_success_new, expected_value_ad, decision in sensitivity_results:
    print(f"Ймовірність успіху рекламної кампанії: {p_success_ad_new}")
    print(f"Вартість успішної рекламної події: {revenue_ad_success_new} тис. дол.")
    print(f"Очікувана вартість з рекламною кампанією: {expected_value_ad} тис. дол.")
    print(f"Рекомендація фірмі: {decision}\n")
```

Зміна параметрів впливає на рішення: зміна двох параметрів - ймовірності успіху рекламної кампанії та вартості успішної рекламної події - суттєво впливає на ваше рішення щодо проведення рекламної кампанії. Значення цих параметрів можуть вирішити, чи доцільно проводити кампанію.
